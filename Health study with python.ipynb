# Requête 1 Marc: Proportion de personnes en état de sous-nutrition

```python
#Importation des librairies 
import pandas as pd
import numpy as np
```

```python
#Importation des données
df_population = pd.read_csv ('population.csv')
df_sous_nutrition = pd.read_csv ('sous_nutrition.csv')
df_dispo_alimentaire = pd.read_csv ('dispo_alimentaire.csv')
df_aide_alimentaire = pd.read_csv ('aide_alimentaire.csv')
```

```python
# Visualisation des data frames
display(df_sous_nutrition.head())
display(df_population.head())
```

```python
#Visualisation du nombre de lignes et colonnes de df_sous_nutrition
display(df_sous_nutrition.shape)
display(df_sous_nutrition.info())

#Visualisation du nombre de lignes et colonnes de df_population
display(df_population.shape)
display(df_population.info())
display(df_sous_nutrition['Valeur'].unique()) #Visualisation des valeurs uniques
```

```python
#Remplacement et affichage des valeurs de l'année 2017 et remplacement des 'nan' et '<0.1' par 0

df_sous_nutrition['Année'] = df_sous_nutrition['Année'].replace(to_replace='2016-2018', value='2017')
df_sous_nutrition['Valeur']= df_sous_nutrition['Valeur'].fillna(0) 
df_sous_nutrition.loc[df_sous_nutrition['Valeur'] == '<0.1', 'Valeur'] = '0'
df_sous_nutrition = df_sous_nutrition.loc[df_sous_nutrition['Année']=='2017']
df_sous_nutrition['Valeur'] = pd.to_numeric(df_sous_nutrition['Valeur'])

display (df_sous_nutrition)
```

```python
#Renomination de colonne: Valeur en Population_sous_nutrition
df_sous_nutrition = df_sous_nutrition.rename(columns={'Valeur':'Population_sous_nutrition'})
display(df_sous_nutrition)
```

```python
#Sélection de l'année 2017 pour avoir seulement les informations 2017
df_population=df_population.loc[df_population["Année"]==2017]
df_population=df_population.rename(columns={'Valeur':'Population'}) #renommage de la colonne Valeur en Population
df_population.head()
```

```python
# Type de données
display(df_sous_nutrition.dtypes)
display(df_population.dtypes)
```

```python
#Jointure pour avoir la population et la population en sous-nutrition dans le même data frame
sous_nutrition=pd.merge(df_population[['Zone','Population']],df_sous_nutrition[['Zone','Population_sous_nutrition']], on='Zone', how='inner')
display(sous_nutrition)
```

```python
#conversion des valeurs de population pour avoir la même unité
sous_nutrition ['Population']=sous_nutrition['Population']*1000
sous_nutrition ['Population_sous_nutrition']=sous_nutrition['Population_sous_nutrition']*1000000
sous_nutrition.inplace=True
sous_nutrition.head()
```

```python
# Agrégation en somme pour avoir population totale
Total_population= sous_nutrition['Population'].sum()
print("Il y a un total de",round(Total_population)," personnes dans le monde")
```

```python
# Agrégation en somme pour avoir population totale en sous-nutrition
Total_population_sous_nutrition=sous_nutrition['Population_sous_nutrition'].sum()
print(round(Total_population_sous_nutrition),"personnes souffrent de sous-nutrition")
```

```python
#Réponse requête 1 : Proportion de la population mondiale en état de sous-nutrition
Proportion_sous_nutrition_mondiale=Total_population_sous_nutrition/Total_population*100
print(round(Proportion_sous_nutrition_mondiale, 2), "% de la population mondiale est en état de sous-nutrition")
```

# Requête 2 Marc: Calcul du nombre théorique de personnes qui pourraient être nourries

```python
#Visulatisation des données 
display(df_dispo_alimentaire.head())
display(df_dispo_alimentaire.info())
```

```python
#Conversion de la disponibilité intérieure en kg
df_dispo_alimentaire['Disponibilité intérieure']=df_dispo_alimentaire['Disponibilité intérieure']*1000000

#Conversion de la 'Disponibilité alimentaire' en (Kcal/personne/an)
df_dispo_alimentaire['Disponibilité alimentaire (Kcal/personne/jour)']=df_dispo_alimentaire['Disponibilité alimentaire (Kcal/personne/jour)']*365
df_dispo_alimentaire.rename(columns={'Disponibilité alimentaire (Kcal/personne/jour)':'Disponibilité alimentaire (Kcal/personne/an)'},inplace=True)
```

```python
#Regroupement par pays
df_dispo_alimentaire_pays= df_dispo_alimentaire.groupby('Zone').sum()
df_dispo_alimentaire_pays.head()
```

```python
# Conversion de la population et Sélection de la zone et de la population
df_population["Population"]=df_population["Population"]*1000
df_population= df_population[["Zone","Population"]]
display(df_population)
```

```python
#Jointure pour enrichir le data frame avec la population (versione simplifiée)

df_proportion_dispo_alimentaire_pays=pd.merge(df_dispo_alimentaire_pays[["Disponibilité alimentaire (Kcal/personne/an)"]],df_population[["Zone","Population"]], on='Zone', how='left')
df_proportion_dispo_alimentaire_pays.head()
```

```python
#Jointure pour enrichir le data frame avec la population et toutes les données
df_proportion_dispo_alimentaire_pays=pd.merge(df_dispo_alimentaire_pays,sous_nutrition[["Zone","Population"]], on='Zone', how='left')
df_proportion_dispo_alimentaire_pays.head()
```

```python
#Calcul pour trouver la quantité totale de Kcal disponible
total_dispo_alimentaire=((df_proportion_dispo_alimentaire_pays['Population'])*
                         df_proportion_dispo_alimentaire_pays['Disponibilité alimentaire (Kcal/personne/an)']).sum()
print("Il y a un total de", total_dispo_alimentaire,"kcal disponible")
```

```python
#Calcul du nombre théorique de personnes pouvante être nourries sur un base d'une moyenne de 2500 kcal/pers/an

besoin_energitique_minimal= 2500
Nombre_jours = 365

nombre_personnes_nourries_theo=total_dispo_alimentaire/(besoin_energitique_minimal*Nombre_jours)
print("Nombre total de personnes pouvant théoriquement être nourries", round(nombre_personnes_nourries_theo),"personnes")
```

```python
# Réponse requête 2 : Calcul de la proportion de la population pouvant être nourrie
proportion_population_nourrie_monde=round(nombre_personnes_nourries_theo/((df_population['Population']).sum())*100,2)
print("La proportion de personnes pouvant théoriquement être nourries est de",proportion_population_nourrie_monde,"%")
```

# Requête 3 Marc: Nombre théorique de personnes qui pourraient être nourries à partir de la disponibilité alimentaire des produits végétaux

```python
#Affichage du data frame df_dispo_alimentaire
display(df_dispo_alimentaire)
```

```python
#Jointure pour avoir un dataframe avec dispo alimentaire et population
df_dispo_alimentaire_vegetale_pays_pop=pd.merge(df_dispo_alimentaire_vegetale_pays,df_population[['Zone','Population']], on='Zone', how='left')
df_dispo_alimentaire_vegetale_pays_pop
```

```python
#Calcul pour trouver la quantité totale de kcal disponible
total_dispo_alimentaire_vegetale_totale=((df_dispo_alimentaire_vegetale_pays_pop['Population'])*
                                         df_dispo_alimentaire_vegetale_pays_pop['Disponibilité alimentaire (Kcal/personne/an)']).sum()
print("Il y a un total de", total_dispo_alimentaire_vegetale_totale,"kcal disponible")
```

```python
#Réponse requête 3 Marc : 
#Calcul de la proportion de personnes pouvant être nourries grâce à la disponibilité végétale
proportion_population_nourrie_monde_vegetale= round(nombre_personnes_nourries_theo_vegetale/
                                                    ((df_population['Population']).sum())*100,2)
print("la proportion de personnes pouvant être nourries grâce à la disponibilité végétale est de", 
      proportion_population_nourrie_monde_vegetale,"%")
```

```python
#On filtre par Nourriture humaine et on en fait la somme
df_cereales_humain= df_cereales_filtre['Élément'].isin(['Nourriture'])
df_cereales_humain= df_cereales_filtre[df_cereales_humain].sum()
print("Il y a une quantité totale de",df_cereales_humain['Quantité en Tonne'], "tonnes de céréales réservée à la nourriture humaine")

#Proportion de céréales pour l'amimentation humaine
Proportion_cereales_humain= df_cereales_humain['Quantité en Tonne']/df_cereales_total['Quantité en Tonne']*100
print("Soit :",round(Proportion_cereales_humain,2),"%")
```

```python
#On filtre par aliments pour animaux et on en fait la somme
df_cereales_animaux= df_cereales_filtre['Élément'].isin(['Aliments pour animaux'])
df_cereales_animaux= df_cereales_filtre[df_cereales_animaux].sum()
print("Il y a une quantité totale de",df_cereales_animaux['Quantité en Tonne'], "tonnes de céréales réservée aux animaux")

#Proportion de céréales pour l'amimentation animale
Proportion_cereales_animaux= df_cereales_animaux['Quantité en Tonne']/df_cereales_total['Quantité en Tonne']*100
print("Soit :",round(Proportion_cereales_animaux,2),"%")
```
